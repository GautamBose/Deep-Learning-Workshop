```python
img = plt.imread('/Users/Gautam/Google Drive/Machine Learning/WORKSHOP/Process/wef.jpg')

plt.imshow(img)

newImg = inception.preprocess(img)

# img.shape
```

```python
tf.reset_default_graph()
tf.import_graph_def(net['graph_def'])
```

```python
gf = tf.get_default_graph()
```

```python
names = [op.name for op in gf.get_operations()]
```

```python
input_name = names[0] + ':0'
X = gf.get_tensor_by_name(input_name)
```

```python
softmax = names[-1] + ':0'
```

```python
sess = tf.InteractiveSession()
```

```python
res = sess.run(softmax, feed_dict = {X:newImg[np.newaxis]})
```

```python
itemIndex = np.argmax(res.mean(axis=0))
```

```python
for label in net['labels']:
    if label[0] == itemIndex:
        print(label[1])
```
